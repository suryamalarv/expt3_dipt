```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histogram equalization for Grayscale images

```python
#Read the "parraot.jpg" image
image=cv2.imread('parrot.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
#Plot the image
plt.imshow(image,cmap='gray')
plt.title('Grayscale image')
plt.show()
```

```python
#To check whether two images are exactly same images
assert np.array_equal(image, cv2.imread('parrot.jpg', cv2.IMREAD_GRAYSCALE))
```

```python
#Display the histogram image
plt.hist(image.ravel(),256,range=[0,256]) #ravel()-flatten the array into 1D view
plt.title('Histogram image')
plt.show()
```

```python
#Equalize the histogram
image_equalization=cv2.equalizeHist(image)
```

```python
#Display the equalized image as histogram
plt.hist(image_equalization.ravel(),256,range=[0,256])
plt.title("Equalized histogram image")
```

```python
#Display the equalization image
plt.imshow(image_equalization,cmap='gray')
plt.title('Equalized image as gray')
plt.show()
```

## Histogram Equalization for color images

```python
# read the color image
color_image=cv2.imread('parrot.jpg',cv2.IMREAD_COLOR)
```

```python
#Convert BGR to HSV
hsv_image=cv2.cvtColor(color_image,cv2.COLOR_BGR2HSV)
```

```python
#Perform histogram equalization only on the V channel, for value intensity.
hsv_image[:,:,2] = cv2.equalizeHist(hsv_image[:, :, 2])
```

```python
#Convert back to BGR format
image_equalization=cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)
```

```python
#Plot the image_equlization
plt.imshow(image_equalization[:,:,::-1])
plt.title('Equalized image')
plt.show()
```

```python
#Plot the image as histogram
plt.hist(image_equalization.ravel(),256,range=[0,256])
plt.title('Histogram equalized image')
plt.show()
```

```python
# Display the original images.
plt.figure(figsize = (8,6))
plt.imshow(color_image[:, :, ::-1])
plt.title('Original Color Image')
plt.show()
```

```python
#Display the equalization image
plt.figure(figsize = (8,6))
plt.imshow(image_equalization[:, :, ::-1])
plt.title('Equalized Image')
plt.show()
```

```python
#Display the color image as histogram
plt.figure(figsize = (8,6))
plt.hist(color_image.ravel(),256,range = [0, 256])
plt.title('Original Image')
plt.show()
```

```python
#Display the equalization image as histogram
plt.figure(figsize = (8,6))
plt.hist(image_equalization.ravel(),256,range = [0, 256])
plt.title('Histogram Equalized')
plt.show()
```
